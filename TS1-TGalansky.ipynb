## Introducción teórica
Una señal analógica continua en el tiempo $x(t)$ puede convertirse en una señal discreta mediante un proceso de muestreo con período $\Delta t$, generando una secuencia. Para que una señal pueda reconstruirse sin ambigüedades, debe cumplirse el criterio de Nyquist, que establece que la frecuencia de muestreo $f_s = 1/\Delta t$ debe ser al menos el doble de la frecuencia máxima presente en la señal.
\begin{equation}
f_s \geq 2 \cdot f_{max}
\end{equation}
Las señales senoidales sirven como modelo ideal para representar fenómenos periódicos. Son de la forma:
\begin{equation}
x(t) = A \sin\big( 2 \pi f t + \phi \big)
\end{equation}
Para desfazar una señal senoidal, se debe modificar el parámetro $\phi$, y para amplificarla se debe modificar la amplitud $A$. La modulación, en cambio, consiste en multiplicar dos señales de la siguiente manera:
\begin{equation}
x(t) = A \sin\big( 2 \pi f_1 t + \phi_1 \big) \cdot B \sin\big( 2 \pi f_2 t + \phi_2 \big)
\end{equation}
Las señales no sinusoidales como la señal cuadrada y el pulso rectangular también son de gran importancia para el análisis y procesamiento digital de señales.
Otro aspecto importante es el cálculo de la potencia y la energía de una señal. Para una señal discreta periódica $x[n]$, la energía se suele considerar la energía de un periodo completo, y se expresa como:
\begin{equation}
E_x = \sum_{n=0}^{T-1} |x[n]|^2
\end{equation}
siendo $T$ el período. La potencia promedio se calcula sumando sobre un período completo, de la siguiente manera:
\begin{equation}
P_x = \frac{1}{T} \sum_{n=0}^{T-1} |x[n]|^2
\end{equation}
La ortogonalidad entre señales también es un concepto fundamental. Dos señales discretas periódicas $x[n]$ e $y[n]$  de longitud $N$ son ortogonales si se cumple la siguiente condición:
\begin{equation}
\sum_{n=0}^{N} x[n] \cdot y[n] = 0
\end{equation}
Si son ortogonales se puede decir que "no comparten información" en ese intervalo. Esto permite representar señales como combinaciones de funciones ortogonales sin interferencia entre ellas, lo cual es importante, por ejemplo, para el análisis de Fourier.

De forma relacionada, la correlación mide el grado de semejanza entre dos señales. Una correlación cruzada normalizada, o un coeficiente de correlación, equivalente a la unidad indica un ajuste perfecto, en tanto que un ajuste deficiente arrojará un valor cercano a cero. [[1]](https://glossary.slb.com/es/terms/c/crosscorrelation) La autocorrelación de una señal $x[n]$ se define como:
\begin{equation}
r_{xx}[m] = \sum_{n=-\infty}^{\infty} x[n] \cdot x[n+m]
\end{equation}
mientras que la correlación cruzada entre $x[n]$ e $y[n]$ es:
\begin{equation}
r_{xy}[m] = \sum_{n=-\infty}^{\infty} x[n] \cdot y[n+m]
\end{equation}
Un resultado interesante que se explorará en este trabajo es la identidad trigonométrica:
\begin{equation}
2 \sin(\alpha)\sin(\beta) = \cos(\alpha - \beta) - \cos(\alpha + \beta)
\end{equation}
la cual se relaciona con la forma en que se pueden descomponer productos de senos en combinaciones de cosenos, lo que es útil en análisis espectral y en la demostración de ortogonalidad de señales sinusoidales.

## Análisis y desarrollo
En primer lugar, se generaron distintas señales en Python, en varios casos tomando como referencia la siguiente señal senoidal de $2 kHz$ de frecuencia y amplitud 1. Además, en cada caso se indica el tiempo entre muestras, el número de muestras y la potencia o la energía según corresponde.

```python
#%%
import numpy as np
import matplotlib.pyplot as plt


def mi_funcion_sen(vmax, dc, ff, ph, nn, fs):
    
    n = np.arange(0,nn)
    tt = n/fs
    w0 = 2 * np.pi * ff
    xx = dc + vmax * np.sin(w0 * tt + ph)
    
    return tt, xx

def mi_funcion_mod(vmax, dc, ff, ph, nn, fs):
    
    n = np.arange(0,nn)
    tt = n/fs
    w0 = 2 * np.pi * ff
    xx = (dc + vmax * np.sin(w0 * tt + ph))*(dc + vmax * np.sin((w0/2) * tt +ph))
    
    return tt, xx

from scipy import signal
def mi_funcion_cuad(vmax=1, dc=0, ff=1, ph=0, nn=1000, fs=1000, duty=0.5):
    tt = np.arange(0,nn)/fs
    xx = signal.square(2 * np.pi * ff * tt + ph, duty=duty)
    xx = dc + (vmax * xx)
    
    return tt, xx
```

```python
#%%
# Primera señal
N = 500
fs1 = 100000  
ts = 1/fs1
vmax1=1
dc1=0
f1=2000
ph1=0
t1,x1 = mi_funcion_sen(vmax=vmax1,dc=dc1,ff=f1,ph=ph1,nn=N,fs=fs1)
Ts1= 1/fs1
p1= (vmax1**2)/2
print("Senoidal de 2 KHz")
print("Tiempo entre muestras: ",Ts1, "s\nNúmero de muestras: ",N, "\nPotencia media: ", p1,"V²")

plt.figure(1)
plt.plot(t1,x1)
plt.title('Senoidal de 2 KHz')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud [V]')
plt.grid(True)
plt.show()
```

Luego, se generó una señal desfazada y amplificada con respecto a ésta.

```python
# Segunda señal
vmax2=2
dc2=0
f2=f1
ph2= np.pi / 2
t2,x2 = mi_funcion_sen(vmax=vmax2,dc=dc2,ff=f2,ph=ph2,nn=N,fs=fs1)
Ts2= 1/fs1
p2= (vmax2**2)/2
print("\nSenoidal amplificada y desfazada")
print("Tiempo entre muestras: ",Ts2, "s\nNúmero de muestras: ",N, "\nPotencia media: ", p2,"V²")

plt.figure(2)
plt.plot(t1,x1)
plt.title('Senoidal amplificada y desfazada')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud [V]')
plt.grid(True)
plt.show()
```

La tercera señal es representa una modulación en amplitud de la primera señal con una igual pero de frecuencia $1 kHz$. Se incluye además un gráfico de ambas señales sin modular a fin de poder realizar una comparación visual.

```python
# Tercera señal
ta,xa = mi_funcion_sen(vmax=vmax1,dc=dc1,ff=f1/2,ph=ph1,nn=N,fs=fs1)
t3,x3 = mi_funcion_mod(vmax=vmax1,dc=dc1,ff=f1,ph=ph1,nn=N,fs=fs1)
Ts3= 1/fs1
p3= (vmax1**2)/2
print("\nSeñal modulada en amplitud")
print("Tiempo entre muestras: ",Ts3, "s\nNúmero de muestras: ",N, "\nPotencia media: ", p3,"V²")

fig, ax = plt.subplots(2, 1)
ax[0].plot(t1, x1, label="Senoidal de 2 KHz", color='blue')
ax[0].plot(ta,xa, label="Senoidal de 1 KHz", color='green')
ax[0].set_title('Senoidales con frecuencia 2KHz y 1KHz')
ax[0].set_xlabel('Tiempo [s]')
ax[0].set_ylabel('Amplitud [V]')
ax[0].legend(loc="upper right")
ax[0].grid(True)

ax[1].plot(t3, x3, color='red')
ax[1].set_title('Señal modulada en amplitud')
ax[1].set_xlabel('Tiempo [s]')
ax[1].set_ylabel('Amplitud [V]')
ax[1].grid(True)

plt.tight_layout()
```

La cuarta señal es la primera recortada al 75% de su amplitud. Esta se generó utilizando la función *clip* de numpy, definiendo el umbral (threshold) anteriormente.

```python
# Cuarta señal
threshold= 0.75 * vmax1
x4 = np.clip(x1,-threshold,threshold)
Ts4= 1/fs1
p4= (threshold**2)/2
print("\nSeñal recortada al 75% de su amplitud")
print("Tiempo entre muestras: ",Ts4, "s\nNúmero de muestras: ",N, "\nPotencia media: ", p4,"V²")
plt.figure(4)
plt.plot(t1,x4)
plt.title('Señal recortada al 75% de su amplitud')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud [V]')
plt.grid(True)
plt.show()
```

La quinta señal, que ya no está relacionada con la primera, es una función cuadrada de $4 kHz$ de frecuencia.

```python
#Quinta señal
t5,x5 = mi_funcion_cuad(vmax=vmax1,dc=dc1,ff=4000,ph=ph1,nn=N,fs=fs1,duty=0.5)
Ts5= 1/fs1
p5= (vmax1**2)/2
print("\nSeñal cuadrada de 4 KHz")
print("Tiempo entre muestras: ",Ts5, "s\nNúmero de muestras: ",N, "\nPotencia media: ", p5,"V²")
plt.figure(5)
plt.plot(t5*2,x5)
plt.title('Señal cuadrada de 4 KHz')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud [V]')
plt.grid(True)
plt.show()
```

La sexta señal es un pulso rectangular de $10 ms$ de duración. Durante el proceso de programación se recurrió al apoyo de herramientas de inteligencia artificial para comprender y aplicar correctamente la función *signal.windows.boxcar* y para ubicar al pulso en el medio del gráfico.

```python
# Sexta señal
fs = 20000       # Frecuencia de muestreo (Hz)
Ts = 1/fs
Tt = 1    # Duración total = 1 segundo
Tp = 0.01        # Duración del pulso = 10 ms

Nt = int(Tt / Ts)   # Muestras totales
Np = int(Tp / Ts)        # Muestras del pulso

# Crear señal con ceros
x6 = np.zeros(Nt)

# Generar pulso rectangular (ventana boxcar de 10ms)
pulse = signal.windows.boxcar(Np)

# Insertar el pulso en el medio de la señal
start = (Nt - Np) // 2
x6[start:start+Np] = pulse

# Vector de tiempo
t6 = np.arange(Nt) * Ts
e6= (vmax1**2)*Tp
print("\nPulso rectangular de 10 ms")
print("Tiempo entre muestras: ",Ts, "s\nNúmero de muestras: ",Nt, "\nEnergía total: ", e6,"V².s")
# Graficar toda la señal (100 ms)
plt.figure(6)
plt.plot(t6*1000, x6)
plt.title("Pulso rectangular de 10 ms")
plt.xlabel("Tiempo [ms]")
plt.ylabel("Amplitud [V]")
plt.grid(True)
plt.show()
```

En segundo lugar, se implementó la función *dot* de numpy, la cual realiza el producto interno de dos señales, para verificar la ortogonalidad de la primera señal con cada una de las demás señales.

```python
#%% Verificar ortogonalidad
dot12 = np.dot(x1, x2)
dot13 = np.dot(x1, x3)
dot14 = np.dot(x1, x4)
dot15 = np.dot(x1, x5)
tolerancia = 1e-10

if abs(dot12) < tolerancia:
    print("x1 y x2 son ortogonales")
else:
    print("x1 y x2 NO son ortogonales")

if abs(dot13) < tolerancia:
    print("x1 y x3 son ortogonales")
else:
    print("x1 y x3 NO son ortogonales")

if abs(dot14) < tolerancia:
    print("x1 y x4 son ortogonales")
else:
    print("x1 y x4 NO son ortogonales")

if abs(dot15) < tolerancia:
    print("x1 y x5 son ortogonales")
else:
    print("x1 y x5 NO son ortogonales")

# Para comparar con el pulso de 10ms necesito que tengan la misma cantidad de muestras
tt1,xx1 = mi_funcion_sen(vmax=vmax1,dc=dc1,ff=f1,ph=ph1,nn=Nt,fs=fs1)
dot16 = np.dot(xx1,x6)
if abs(dot16) < tolerancia:
    print("x1 y x6 son ortogonales")
else:
    print("x1 y x6 NO son ortogonales")
```

Se consideró una cierta tolerancia, ya que en la práctica es difícil que el producto interno se anule exactamente. La señal original $x_1$ resulta ortogonal con $x_2$, ya que se trata de la misma frecuencia con un desfase de $\pi/2$, lo cual garantiza que el producto se anule siempre que el cálculo se realice sobre un número entero de periodos. También es ortogonal con $x_3$, ya que el producto interno resulta ser entre tres funciones senos, dos con frecuencia $2 kHz$ y el otro con frecuencia $1 kHz$, siendo esta última la que proporciona "información" que la otra no tiene. Esto también se puede inferir observando el gráfico comparativo de las señales.

Por otro lado, la señal recortada $x_4$ no es ortogonal con $x_1$, dado que las señales son proporcionales. Recortar una señal no hace ortogonal a algo que antes no lo era.

En el caso de la señal cuadrada $x_5$ de $4 kHz$, es ortogonal con $x_1$ dado que, de manera similar al caso de $x_3$, su desarrollo en serie de Fourier no contiene la frecuencia de $2 kHz$ y, bajo la condición de integrar un periodo común entero, no hay solapamiento en frecuencia.

Finalmente, el caso del pulso rectangular $x_6$ es ortogonal, ya que el pulso abarca un número entero de periodos completos de la senoidal de $2 kHz$ y está alineado con las muestras, y, por lo tanto, la suma de productos se anula.

En tercer lugar, se graficó la autocorrelación de $x_1$ y su correlación cruzada con las demás señales.

```python
#%% Autocorrelacion y correlacion cruzada
def correlacion(x, y):
    corr = np.correlate(x, y, mode='full') * ts
    lags = np.arange(-len(x)+1, len(x)) * ts
    return corr, lags

auto, lags1 = correlacion(x1,x1)
corr12, lags2 = correlacion(x1,x2)
corr13, lags3 = correlacion(x1,x3)
corr14, lags4 = correlacion(x1,x4)
corr15, lags5 = correlacion(x1,x5)

# Primero asegurar que x1 tenga la misma longitud que x6 (20,000 muestras)
t_total = np.arange(0, Tt, Ts)  # 1 segundo, 20,000 muestras
x1_extendida = np.sin(2 * np.pi * f1 * t_total)  # Sinusoidal completa de 1 segundo

# Ahora sí recortar ambas señales en el mismo intervalo
indices = np.where(x6 != 0)[0]
start, end = indices[0], indices[-1]

x6_recortada = x6[start:end+1]
x1_recortada = x1_extendida[start:end+1]  # Ahora sí tienen la misma longitud
corr16, lags6 = correlacion(x1_recortada,x6_recortada)


# Graficos
lags = np.arange(-len(x1)+1, len(x1))
fig, axs = plt.subplots(3, 2, figsize=(10,8))

axs[0,0].plot(lags1, auto, color='blue')
axs[0,0].set_title("Autocorrelación de x1 (2 kHz)")
axs[0,0].set_xlabel('Demora [s]')
axs[0,0].set_ylabel('Correlación')
axs[0,0].grid(True)

axs[0,1].plot(lags2, corr12, color='red')
axs[0,1].set_title("Correlación x1 vs x2")
axs[0,1].set_xlabel('Demora [s]')
axs[0,1].set_ylabel('Correlación')
axs[0,1].grid(True)

axs[1,0].plot(lags3, corr13, color='green')
axs[1,0].set_title("Correlación x1 vs x3")
axs[1,0].set_xlabel('Demora [s]')
axs[1,0].set_ylabel('Correlación')
axs[1,0].grid(True)

axs[1,1].plot(lags4, corr14, color='purple')
axs[1,1].set_title("Correlación x1 vs x4")
axs[1,1].set_xlabel('Demora [s]')
axs[1,1].set_ylabel('Correlación')
axs[1,1].grid(True)

axs[2,0].plot(lags5, corr15, color='orange')
axs[2,0].set_title("Correlación x1 vs x5")
axs[2,0].set_xlabel('Demora [s]')
axs[2,0].set_ylabel('Correlación')
axs[2,0].grid(True)

axs[2,1].plot(lags6, corr16, color='magenta')
axs[2,1].set_title("Correlación x1 vs x6")
axs[2,1].set_xlabel('Demora [s]')
axs[2,1].set_ylabel('Correlación')
axs[2,1].grid(True)
# Ajustar diseño
plt.tight_layout()
plt.show()
```

El análisis de correlación revela propiedades fundamentales de las señales. La autocorrelación de la sinusoidal de $2kHz$ presenta un pico máximo en 0 y decaimiento periódico, confirmando su periodicidad. La correlación cruzada con $x_2$ muestra un comportamiento cosenoidal (provocado por el desfazamiento) con valor medio nulo, verificando su ortogonalidad. La correlación con la señal modulada $x_3$ resulta no nula y no periódica pura, y presenta múltiples componentes frecuenciales.

Se puede observar que la única diferencia entre correlación con $x_4$ y la autocorrelación de $x_1$ es la amplitud, lo cual es esperado ya que las señales tienen esencialmente los mismos parámetros.
Por otro lado, la correlación con la señal cuadrada $x_5$ presenta una forma compleja con múltiples componentes armónicas superpuestas. Se observa una oscilación principal a $4kHz$ modulada por contribuciones de armónicos impares superiores ($12kHz$, $20kHz$, etc.), resultado de la naturaleza discontinua de la señal cuadrada. La correlación cruzada con el pulso rectangular $x_6$ conserva la frecuencia de la senoidal $2 kHz$ y su amplitud depende de la coincidencia del pulso rectangular con las zonas positivas o negativas de la senoidal. Por eso se ven regiones con valores positivos (cuando se solapan los semiciclos positivos) y negativos (cuando coinciden con los semiciclos negativos), manteniendo la simetría par típica de la correlación cruzada.

Estas observaciones confirman que la autocorrelación mide la similitud temporal de una señal consigo misma, mientras que la correlación entre señales permite cuantificar la semejanza y la ortogonalidad entre distintas señales, tal como se predice mediante la identidad trigonométrica que relaciona productos de senos con sumas de cosenos. El siguiente código realiza un análisis fundamental de una señal de audio en formato WAV, para el cual se eligió un audio de una tijera cortando pelo, comenzando con la lectura del archivo para obtener la frecuencia de muestreo y los datos de audio, los cuales son normalizados al rango $[-1, 1]$ para garantizar consistencia en el procesamiento. Luego, grafica la amplitud de la señal contra su duración, lo que permite observar las variaciones temporales del sonido. Finalmente, calcula la energía total de la señal mediante la suma de los cuadrados de todas las muestras.

```python
#%% Bonus
from scipy.io import wavfile

filename = "audio_tijeras.wav"

# Lectura del archivo
fs, audio_data = wavfile.read(filename)

# Normalizar a valores entre -1 y 1
audio_data = audio_data.astype(np.float32) / np.max(np.abs(audio_data))

# Grafico
plt.figure(figsize=(12, 6))
time = np.arange(len(audio_data)) / fs
plt.plot(time, audio_data)
plt.title(f"Forma de onda del audio: {filename}")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()

# cálculo de la energía
# energía = suma de los cuadrados de las muestras
energia = np.sum(audio_data**2)

print("Análisis del archivo de audio")
print("Frecuencia de muestreo:",fs," Hz")
print("Duración:",len(audio_data)/fs," segundos")
print("Número de muestras:", len(audio_data))
print("Energía total:", energia," V²·muestras")
```

En cambio, la señal recortada no es ortogonal, pues solo reduce amplitud sin modificar su dependencia. En el caso del pulso rectangular de $10 ms$, la ortogonalidad se cumplió al abarcar ciclos completos que permiten la cancelación integral de áreas.

La señal cuadrada con duty 50% presentó una potencia de $0.5 V^2$, correspondiente al promedio de su amplitud, mientras que la modulación implementada como producto directo de senoidales mostró una potencia coherente con la redistribución espectral descrita por la identidad trigonométrica $2 \sin(\alpha)\sin(\beta) = \cos(\alpha - \beta) - \cos(\alpha + \beta)$. Esta compatibilidad con los valores esperados de potencia se deben a que se eligieron correctamente los valores de la frecuencia de muestreo, la cantidad de muestras y la duración de la ventana de análisis. El pulso rectangular confirmó su naturaleza de energía finita ($0.01 V^2·s$), distinguiéndose claramente de las señales periódicas.

El análisis de correlación mostró que la autocorrelación de la senoidal refleja su periodicidad, mientras que las correlaciones cruzadas evidencian similitud y diferencias entre señales: el desfase genera una correlación cosenoidal, la modulación incorpora múltiples frecuencias, la cuadrada introduce armónicos y el pulso rectangular conserva la frecuencia original con inversión de signo según el solapamiento. La identidad trigonométrica permitió interpretar estas relaciones como redistribuciones espectrales del producto de senoidales.

La normalización asegura que el valor refleje la energía relativa de la señal, validando la metodología empleada para su cálculo.
