## Introducción
Las señales senoidales sirven como modelo ideal para representar fenómenos periódicos. Se pueden describir completamente a partir de ciertos parámetros en el dominio temporal: su amplitud máxima, su valor medio, la frecuencia, y la fase. 

En un entorno digital, las señales continuas deben discretizarse mediante un proceso de muestreo, llevado a cabo por un convertidor analógico-digital (ADC), que transforma la señal continua en una serie de muestras numéricas. Un filtro pasa bajos se aplica para recuperar la señal original a partir de sus muestras, eliminando las copias del espectro que aparecen debido al muestreo. De esta manera, se preserva únicamente el contenido original de la señal, siempre que se cumpla el criterio de Nyquist.

El teorema del muestreo de Nyquist-Shannon establece que, para poder reconstruir correctamente una señal analógica a partir de sus muestras digitales, la frecuencia de muestreo $f_s$ debe ser al menos el doble de la frecuencia máxima presente en la señal $(f_{max})$.
\begin{equation}
f_s \geq 2 \cdot f_{max}
\end{equation}
Cuando no se cumple esta condición, ocurre un fenómeno denominado *aliasing* en el cual la señal reconstruida se distorsiona como consecuencia de un submuestreo. Por lo tanto, se deben tomar en cuenta como parámetros adicionales la cantidad de muestras digitalizada por el ADC y la frecuencia de muestreo del ADC.

Una fórmula habitual para calcular la frecuencia aliaseda es: $f_{alias}= f - k . f_s$ donde $k$ es el entero que aproxima $f$ al múltiplo más cercano de $f_s$. 

## Análisis y desarrollo
En primer lugar, se programó en Python una función que genera señales senoidales con la posibilidad de modificar su amplitud máxima ($vmax$), su valor medio ($dc$), la frecuencia ($ff$), la fase ($ph$), la cantidad de muestras ($nn$) y $f_s$. Se muestra un ejemplo con una frecuencia de 2 Hz.

```python
import numpy as np
import matplotlib.pyplot as plt


def mi_funcion_sen(vmax, dc, ff, ph, nn, fs):
    
    n = np.arange(0,nn)
    tt = n/fs
    w0 = 2 * np.pi * ff
    xx = dc + vmax * np.sin(w0 * tt + ph)
    
    return tt, xx
```

```python
N = 1000
fs = N
vmaxs=1
dcs=0
fas=0
tt,xx = mi_funcion_sen(vmax=vmaxs,dc=dcs,ff=2,ph=fas,nn=N,fs=fs)

plt.figure(1)
plt.plot(tt,xx)
plt.title("Señal senoidal")
plt.xlabel("Tiempo [s]"), plt.ylabel("Amplitud [V]")
plt.grid(True)
plt.show()
```

## Bonus
Se realizaron distintos experimentos variando la frecuencia de la señal senoidal y observando su representación bajo una frecuencia de muestreo fija ($f_s = 1000$).

```python
tt,xx = mi_funcion_sen(vmax=vmaxs,dc=dcs,ff=500,ph=fas,nn=N,fs=fs)
plt.figure(2)
plt.plot(tt,xx)
plt.title("Señal senoidal de frecuencia 500 Hz")
plt.xlabel("Tiempo [s]"), plt.ylabel("Amplitud [V]")
plt.grid(True)
plt.show()

tt,xx = mi_funcion_sen(vmax=vmaxs,dc=dcs,ff=999,ph=fas,nn=N,fs=fs)
plt.figure(3)
plt.plot(tt,xx)
plt.title("Señal senoidal de frecuencia 999 Hz")
plt.xlabel("Tiempo [s]"), plt.ylabel("Amplitud [V]")
plt.grid(True)
plt.show()

tt,xx = mi_funcion_sen(vmax=vmaxs,dc=dcs,ff=1001,ph=fas,nn=N,fs=fs)
plt.figure(4)
plt.plot(tt,xx)
plt.title("Señal senoidal de frecuencia 1001 Hz")
plt.xlabel("Tiempo [s]"), plt.ylabel("Amplitud [V]")
plt.grid(True)
plt.show()

tt,xx = mi_funcion_sen(vmax=vmaxs,dc=dcs,ff=2001,ph=fas,nn=N,fs=fs)
plt.figure(5)
plt.plot(tt,xx)
plt.title("Señal senoidal de frecuencia 2001 Hz")
plt.xlabel("Tiempo [s]"), plt.ylabel("Amplitud [V]")
plt.grid(True)
plt.show()
```

Estos resultados son interesantes para comprender mejor el fenómeno de aliasing.

En primer lugar, se observa que la frecuencia de 500 Hz cumple exactamente $f_s=2 f_{max}$, por lo tanto estamos realizando un muestreo crítico. La señal aparece distorsionada, tomando una forma que recuerda a un árbol de Navidad de costado. Esto ocurre porque para reconstruir la señal se necesitaría un filtro pasa bajo ideal, y éste es irrealizable en la práctica.

En cambio, para la frecuencia de 999 Hz, la señal muestreada se ve como una senoidal invertida respecto de la esperada. Al muestrear periódicamente una señal senoidal, puede suceder que las muestras coincidan con las que generaría otra senoidal de menor frecuencia. En particular, si para una senoidal no se cumple el criterio de Nyquist, las muestras obtenidas serán indistinguibles de las de una senoidal de frecuencia $f - f_s$. Por eso, en este caso la frecuencia resultante del aliasing es $999 Hz - 1000 Hz = -1 Hz$.

En los casos de frecuencia de 1001 Hz y 2001 Hz, las señales recuperan la apariencia de una senoidal “normal”. Análogamente, no se trata de la frecuencia real, sino de una frecuencia aliaseada que cae dentro de la banda permitida por el teorema de Nyquist, ya que estas son $1001 Hz - 1000 Hz = 1 Hz$ y $2001 Hz - 2 . 1000 Hz = 1 Hz$.

```python
from scipy import signal
def mi_funcion_cuad(vmax=1, dc=0, ff=1, ph=0, nn=1000, fs=1000, duty=0.5):
    tt = np.arange(0,nn)/fs
    xx = signal.square(2 * np.pi * ff * tt + ph, duty=duty)
    xx = dc + (vmax * xx)
    # Asegurar forma Nx1
    tt = tt.reshape(-1,1)
    xx = xx.reshape(-1,1)
    
    return tt, xx
```

```python
N = 1000
fs = 1000
tt, xx = mi_funcion_cuad(vmax=1, dc=0, ff=5, ph=0, nn=N, fs=fs, duty=0.7)

plt.plot(tt, xx)
plt.xlabel("Tiempo [s]"), plt.ylabel("Amplitud [V]")
plt.title("Señal cuadrada")
plt.grid(True)
plt.show()
```

Luego, se implementó una segunda función para generar señales cuadradas. En este caso, se aprovechó la relación entre una onda senoidal y la función sign, de manera de obtener una señal binaria con valores $± V_{max}$, pudiendo además incluir un valor medio (offset) y modificar el duty cycle de la señal. Durante el proceso de programación se recurrió al apoyo de herramientas de inteligencia artificial para comprender y aplicar correctamente la función sign.
